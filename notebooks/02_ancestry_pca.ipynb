```python
# Cell 1 — visualize 1000 Genomes PCA colored by population
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
%matplotlib inline

# Load PCA results
eigenvec = pd.read_csv(
    "/home/bmurnyak/cancer-ancestry-genomics/data/reference/1kg_chr22_pca.eigenvec",
    sep='\t', header=0)
eigenval = pd.read_csv(
    "/home/bmurnyak/cancer-ancestry-genomics/data/reference/1kg_chr22_pca.eigenval",
    header=None, names=['eigenvalue'])

# Load population panel
panel = pd.read_csv(
    "/home/bmurnyak/cancer-ancestry-genomics/data/reference/1kg_samples.panel",
    sep='\t')

print("Eigenvec columns:", eigenvec.columns.tolist()[:5])
print("Panel columns:", panel.columns.tolist())
print(f"\nSamples in PCA: {len(eigenvec)}")
print(f"Samples in panel: {len(panel)}")
print(f"\nSuperpopulations:")
print(panel['super_pop'].value_counts())

# Variance explained
total = eigenval['eigenvalue'].sum()
var_explained = (eigenval['eigenvalue'] / total * 100).round(2)
print(f"\nVariance explained by PC1: {var_explained[0]:.2f}%")
print(f"Variance explained by PC2: {var_explained[1]:.2f}%")
```

```python
# Cell 2 — 1000 Genomes PCA plot colored by superpopulation
# Merge PCA with population info
pca_df = eigenvec.merge(panel, left_on='IID', right_on='sample', how='left')

superpop_colors = {
    'AFR': '#E24B4A',  # African
    'EAS': '#1D9E75',  # East Asian
    EUR  : '#185FA5',  # European
    'SAS': '#BA7517',  # South Asian
    'AMR': '#534AB7'   # American
}
superpop_colors = {
    'AFR': '#E24B4A',
    'EAS': '#1D9E75',
    'EUR': '#185FA5',
    'SAS': '#BA7517',
    'AMR': '#534AB7'
}

superpop_labels = {
    'AFR': 'African (n=661)',
    'EAS': 'East Asian (n=504)',
    'EUR': 'European (n=503)',
    'SAS': 'South Asian (n=489)',
    'AMR': 'Admixed American (n=347)'
}

fig, axes = plt.subplots(1, 2, figsize=(16, 7))

for ax, (pc_x, pc_y) in zip(axes, [('PC1','PC2'), ('PC1','PC3')]):
    for spop, group in pca_df.groupby('super_pop'):
        ax.scatter(group[pc_x], group[pc_y],
                   c=superpop_colors[spop],
                   label=superpop_labels[spop],
                   s=15, alpha=0.6, edgecolors='none')
    
    ax.set_xlabel(f'{pc_x} ({var_explained[int(pc_x[-1])-1]:.1f}%)', fontsize=11)
    ax.set_ylabel(f'{pc_y} ({var_explained[int(pc_y[-1])-1]:.1f}%)', fontsize=11)
    ax.set_title(f'{pc_x} vs {pc_y}', fontweight='bold', fontsize=12)
    ax.legend(fontsize=8, markerscale=2)
    ax.grid(alpha=0.3)

plt.suptitle('1000 Genomes Phase 3 — Population Structure PCA\nChr22, 110,401 SNPs, n=2,504 samples',
             fontsize=13, fontweight='bold')
plt.tight_layout()
plt.savefig('/home/bmurnyak/cancer-ancestry-genomics/results/plots/1kg_pca.png',
            dpi=150, bbox_inches='tight')
plt.show()
print("✓ 1000 Genomes PCA saved")
```
